```python
import pandas as pd
import numpy as np
import sqlite3
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("netflix data C W project.csv")
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
# Drop rows with no title or release year
df = df.dropna(subset=['title', 'releaseYear'])
```

```python
# Fill missing rating with a placeholder
df['rating'] = df['imdbAverageRating'].fillna('Not Rated')
df.drop("imdbAverageRating" , axis = 1 )
```

# Handling Missing IMDb Ratings

Many entries in the dataset have missing IMDb ratings. To ensure consistency and avoid issues during analysis or visualization, these missing values were replaced with the placeholder `'Not Rated'`. This makes it easier to identify unrated titles during analysis without excluding them.

After replacement, the original `imdbAverageRating` column was dropped to keep the dataset clean and avoid redundancy, since all necessary info is now stored in the new `rating` column.

This step is crucial in:
- Preserving all records, even those with incomplete data
- Making future groupings and filtering based on ratings easier
- Ensuring that rating-related visualizations (like bar charts) include the `"Not Rated"` category


```python
df = df.dropna(subset=['imdbNumVotes'])
df
```

```python
# Normalize genres (listed_in) to a list format
df['genres'] = df['genres'].apply(lambda x: [genre.strip('') for genre in x.split(',')] if isinstance(x, str) else x)
df = df.dropna(subset=['genres'])
df.head(10)
```

```python
# Normalize country availability to a list
df['countries'] = df['availableCountries'].fillna('').apply(lambda x: [country.strip() for country in x.split(',') if country.strip()])
df.head(10)
```

#  Normalize Genre Information

The `genres` column (previously `listed_in`) contains multiple genres in a single string, separated by commas. To enable easier filtering, analysis, and visualization, we normalize this column by:
- Splitting the string into a list of individual genres
- Stripping extra spaces from each genre
- Dropping rows with missing genre data

This results in a structured `genres` column where each entry is a list of clean genre names.


```python
# Drop duplicates based on show_id
df = df.drop_duplicates(subset=['imdbId'])
df.head(10)
```

```python
df.dtypes
```

```python
# Clean column names to snake_case for SQL
df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]
df.head(10)
```
